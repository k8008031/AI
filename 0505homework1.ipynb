```python
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx')#在(0,0)等4個地方g色x標記
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--')
plt.axis([0,5,0,1])#寬度由0到5 高度由0到1
plt.ylabel('y')
plt.xlabel('x')
plt.legend(('price','passenger'),loc='upper right')
plt.show()
```
