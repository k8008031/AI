```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn  import datasets
from sklearn.model_selection import train_test_split  #匯入kMEAN函式庫
from sklearn.cluster  import KMeans
from sklearn  import metrics


#取得資料
iris=datasets.load_iris()

#拆分資料80:20 訓練與測試用 2類標籤答案
iris_X_train,iris_X_test,iris_y_train,iris_y_test=train_test_split(iris.data,iris.target,test_size=0.2)

#kmeans演算法
kmeans=KMeans(n_clusters=3)
kmeans.fit(iris_X_train)
y_predict=kmeans.predict(iris_X_train)



x1=iris_X_train[:,0]#花萼長度
y1=iris_X_train[:,1]#花萼寬度
plt.scatter(x1,y1,c=y_predict,cmap="viridis")#畫每一筆資料
centers=kmeans.cluster_centers_#每個分類的中心點
plt.scatter(center[:,0],center[:,1],c="black",s=200,alpha=0.5);#畫出中心點
plt.show()
```
